**Seasonal pattern in sales**

```python
merged_orders['date'] = pd.to_datetime(merged_orders['date'])
```

```python
(
    merged_orders
    .resample("M", on="date")
    .agg({'price': 'sum',
          'unit_price': 'sum'})
    .rename(columns={'price': 'revenue generated from original price',
                    'unit_price': 'revenue generated from discounted price'})
    .plot(figsize=(7, 5))
 );
```

First, to see the revenue behaviours from the real situation when the products are sold with discounted applied (orange line) compared to if the products were sold on the full original price (blue line). The difference between the 2 lines can represent the lost revenue from the discount. Althoug, the lost revenue is biggest between the 2 peaks in Nov/Dec time, the total revenue is also by far the highest.

```python
(
    merged_orders
    .resample("M", on="date")
    .agg({'diff_price': 'sum'})
    .rename(columns={'diff_price': 'total discounts'})
    .plot(figsize=(7, 5))
 );
```

second, the zoom in of the lost revenue from the discount (the difference between the real sold price with discount & the originial based price). It can be seen there is a peak it lost revenue in Nov/Dec period. It it because of there are higher discounts given during the time?

```python
(
    merged_orders
    .resample("M", on="date")
    .agg({'%_discount': 'mean'})
    .rename(columns={'%_discount': 'average discount'})
    .plot(figsize=(7, 5))
 );
```

It turns out that, the 'percentage' of given discount is not directly related to the high lost revenue during this period. Then if it is not because of higher percentage of the discount, the massive lower revenue could be explained by the higher number of products sold.

```python
(
    merged_orders
    .resample("M", on="date")
    .agg({'product_quantity': 'sum'})
    .rename(columns={'product_quantity': 'total number of product sold'})
    .plot(figsize=(7, 5))
 );
```

Here the product quatity correlate with the high numbers of products sold, which explains the high lost revenue during the Nov/Dec period. However, the reason of it could still be investigated further. For example, Is if because, even though the % of discount is not different than other season, there are 'more number of prodcuts given discount'?
